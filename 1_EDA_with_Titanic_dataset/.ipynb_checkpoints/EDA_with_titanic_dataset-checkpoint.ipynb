```python
# Import the pandas library
import pandas as pd
```

```python
# Load it into the dataframe
df = pd.read_csv("train.csv")
df.head()
```

```python
# Data Exploration
# 1. Checking data types and nulls
df.info()
```

```python
# 2. Summary statistics
df.describe()
```

```python
# 3. Counting missing values per column
df.isnull().sum()
```

```python
df.columns.tolist()
df.shape
df['Survived'].value_counts(normalize=True)
```

```python
# Data Cleaning
# Handling missing values

# 1. Age - Filling with median value
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
# 2. Embarked - Filling with mode
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
# 3. Cabin - dropping the column as too many missing values are present
df.drop('Cabin', axis=1, inplace=True)
```

```python
# Droppping irrelavent columns
df.drop(['Name', 'Ticket', 'PassengerId'], axis=1, inplace=True)
```

```python
# Encoding Categorial Variable

# using get_dummies for 'Sex' and 'Embarked'
df = pd.get_dummies(df, columns=['Sex', 'Embarked'], drop_first=True)
```

```python
# Data Visualization - EDA

#Importing required libraries
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
# 1. Survival by Gender

sns.countplot(x='Survived', hue='Sex_male', data=df)
plt.title("Survival by Gender")
plt.xticks([0, 1], ['Not Survived', 'Survived'])
plt.show()
```

```python
# 2. Survival by Pclass

sns.countplot(x='Survived', hue='Pclass', data=df)
plt.title("Survival by Passenger Class")
plt.xticks([0, 1], ['Not Survived', 'Survived'])
plt.show()
```

```python
# 3. Age distribution

sns.histplot(data=df, x='Age', hue='Survived', bins=30, kde=True)
plt.title("Age Distribution by Survival")
plt.show()
```

```python
# 4. Heatman of correlation

plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# Saving the cleaned dataset

df.to_csv("cleaned_titanic.csv", index=False)
```
